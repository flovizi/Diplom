#### 2.3 Вспомогательные функции

```python
class class_prediction:
    
    def _init_(self):
        self= self
        
    def get_model(self, X_train, X_test, y_train, y_test, model, random_seed=0):
        self.X_train = X_train
        self.y_train = y_train
        self.X_test = X_test
        self.y_test = y_test
        self.model = model
        
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        self.y_pred = y_pred
        return model

    # Метрики качества    
    def MSE_mean_squared_error(self):
        return mean_squared_error(self.y_test, self.y_pred)

    def MAE_mean_absolute_error(self):
        return mean_absolute_error(self.y_test, self.y_pred)

    def score(self):
        return self.model.score(self.X_test, self.y_test)

    def cross_val(self, cv=KFold(n_splits=10, shuffle=True), n_jobs=-1):
        return cross_val_score(self.model, self.X_train, self.y_train, cv=cv, n_jobs=n_jobs).mean()
    
    def plot(self):
        plt.figure(figsize=(5,4))
        sns.set_style('dark')
        plot = sns.scatterplot(self.y_test, self.y_pred)
        plot.set(xlabel='Given', ylabel='Prediction')
        x_plot = np.linspace(-1000,60000,100)
        y_plot = x_plot
        plt.plot(x_plot, y_plot, color='r')
        plt.grid()
        plt.show()
        
    def leftovers(self):
        plt.scatter(self.model.predict(self.X_train),  self.model.predict(self.X_train) - self.y_train,
            c='blue', marker='o', label='Training data')
        plt.scatter(self.model.predict(self.X_test),  self.model.predict(self.X_test) - self.y_test,
            c='lightgreen', marker='s', label='Test data')
        plt.xlabel('Predicted values')
        plt.ylabel('Residuals')
        plt.legend(loc='upper left')
        plt.hlines(y=0, xmin=-6000, xmax=60000, lw=2, color='red')
        plt.xlim([-6000, 60000])
        plt.tight_layout()
```

###### 4.1.1 Бакетинг  
Это техника обработки данных, при которой исходный набор данных делится на некоторое количество интервалов, общее значение которых затем представляет все значения интвервала, что позволяет уменьшить влияние отдельных наблюдений. Иными словами мы переводим непрерывную величину в дискретную.

Применим бакетинг (биннинг) к признакам 'age' и 'bmi' как добавление признака 

```python
# age
df_X['age_bins'] = np.floor_divide(np.asarray(df['age']), 10)
```

bins_bmi:
* 'deficit': bmi < 18.5   
* 'norma': bmi <= 25   
* 'excess': bmi <= 30    
* 'obesity': bmi > 30  

```python
def bins(x):
    if x<18.5:
        x=1
    elif x<=25:
        x=2
    elif x<=30:
        x=3
    elif x>30:
        x=4
    return x
```

#### 6.3 Стекинг  StackingRegressor

Многоуровневая регрессия.  
Индивидуальные регрессионные модели обучаются на основе полного набора обучающих данных;  
затем устанавливается мета-регрессор на основе выходных данных этих отдельных регрессионных моделей.
